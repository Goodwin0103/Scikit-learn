```python
# Daten säubern
from sklearn.preprocessing import StandardScaler
X_train = Data.drop(columns = ['value'])
y_train = Data['value']
```

<a id='build_model'></a><div><img src="Images/IMG-diagram.png" style="float:left"> <h2 style="position: relative; top: 6px; left:10px">4. Modellauswahl</h2>
<p style="position: relative; top: 10px">
Trainieren Sie ein Regressionsmodell zur Vorhersage des Werts einer Immobilie. Denken Sie unabhängig von der Art des gewählten Modells daran, alle Hyperparameter zu optimieren.

```python
# from sklearn.linear_model import LinearRegression
import numpy as np # Für die Matrixoperationen
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import RidgeCV
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
###############################################
#############Linear Regression#################
##############################################

#################################################################################
stdScaler = StandardScaler().fit(X_train)
X_train = stdScaler.transform(X_train)

X_2_train = np.concatenate([X_train, X_train ** 2], axis=1)
stdScaler_2 = StandardScaler().fit(X_2_train)
X_2_train = stdScaler_2.transform(X_2_train)

X_3_train = np.concatenate([X_train, X_train ** 2, X_train ** 3], axis=1)
stdScaler_3 = StandardScaler().fit(X_3_train)
X_3_train = stdScaler_3.transform(X_3_train)

X_4_train = np.concatenate([X_train, X_train ** 2, X_train ** 3,X_train ** 4], axis=1)
stdScaler_4 = StandardScaler().fit(X_4_train)
X_4_train = stdScaler_4.transform(X_4_train)

X_5_train = np.concatenate([X_train, X_train ** 2, X_train ** 3, X_train ** 4, X_train ** 5], axis=1)
stdScaler_5 = StandardScaler().fit(X_5_train)
X_5_train = stdScaler_5.transform(X_5_train)


X_6_train = np.concatenate([X_train, X_train ** 2, X_train ** 3, 
                            X_train ** 4, X_train ** 5, X_train ** 6], axis=1)
stdScaler_6 = StandardScaler().fit(X_6_train)
X_6_train = stdScaler_6.transform(X_6_train)

X_7_train = np.concatenate([X_train, X_train ** 2, X_train ** 3, 
                            X_train ** 4, X_train ** 5, X_train ** 6,
                           X_train ** 7], axis=1)
stdScaler_7 = StandardScaler().fit(X_7_train)
X_7_train = stdScaler_7.transform(X_7_train)

X_8_train = np.concatenate([X_train, X_train ** 2, X_train ** 3, 
                            X_train ** 4, X_train ** 5, X_train ** 6,
                           X_train ** 7, X_train ** 8], axis=1)
stdScaler_8 = StandardScaler().fit(X_8_train)
X_8_train = stdScaler_8.transform(X_8_train)
#################################################################################

lr1_model = LinearRegression()
lr1_model.fit(X_train, y_train)
train_score_lr1 = lr1_model.score(X_train, y_train)
print("Lineare Regression 1. Ordnung - Training: R² = {:.2f}".format(train_score_lr1))

############################ ergänzt Merkmale 2. Ordnung ###########################

lr2_model =  LinearRegression()
lr2_model.fit(X_2_train, y_train)
train_score_lr2 = lr2_model.score(X_2_train, y_train)
N = X_2_train.shape[0]
M = X_2_train.shape[1]
R2_korr = 1 - (1-train_score_lr2) * (N-1) / (N-M-1)
print("Das korrigerte R² für die Trainingsdaten beträgt {:.2f}.".format(R2_korr))

###########################ergänzt Merkmale 3. Ordnung###########################


lr3_model =  LinearRegression()
lr3_model.fit(X_3_train, y_train)
train_score_lr3 = lr3_model.score(X_3_train, y_train)
N = X_3_train.shape[0]
M = X_3_train.shape[1]
R2_korr = 1 - (1-train_score_lr3) * (N-1) / (N-M-1)
print("Lineare Regression 3. Ordnung - Training: R² = {:.2f}".format( R2_korr ))

###########################ergänzt Merkmale 4. Ordnung###########################


lr4_model =  LinearRegression()
lr4_model.fit(X_4_train, y_train)
train_score_lr4 = lr4_model.score(X_4_train, y_train)
N = X_4_train.shape[0]
M = X_4_train.shape[1]
R2_korr = 1 - (1-train_score_lr4) * (N-1) / (N-M-1)
print("Lineare Regression 4. Ordnung - Training: R² = {:.2f}".format( R2_korr ))

###########################ergänzt Merkmale 5. Ordnung###########################


lr5_model =  LinearRegression()
lr5_model.fit(X_5_train, y_train)
train_score_lr5 = lr5_model.score(X_5_train, y_train)
N = X_5_train.shape[0]
M = X_5_train.shape[1]
R2_korr = 1 - (1-train_score_lr5) * (N-1) / (N-M-1)
print("Lineare Regression 5. Ordnung - Training: R² = {:.2f}".format( R2_korr ))

##########################ergänzt Merkmale 6. Ordnung###########################


lr6_model =  LinearRegression()
lr6_model.fit(X_6_train, y_train)
train_score_lr6 = lr6_model.score(X_6_train, y_train)
N = X_6_train.shape[0]
M = X_6_train.shape[1]
R2_korr = 1 - (1-train_score_lr6) * (N-1) / (N-M-1)
print("Lineare Regression 6. Ordnung - Training: R² = {:.2f}".format( R2_korr ))

##########################ergänzt Merkmale 7. Ordnung###########################


lr7_model =  LinearRegression()
lr7_model.fit(X_7_train, y_train)
train_score_lr7 = lr7_model.score(X_7_train, y_train)
N = X_7_train.shape[0]
M = X_7_train.shape[1]
R2_korr = 1 - (1-train_score_lr7) * (N-1) / (N-M-1)
print("Lineare Regression 7. Ordnung - Training: R² = {:.2f}".format( R2_korr ))

##########################ergänzt Merkmale 8. Ordnung###########################


lr8_model =  LinearRegression()
lr8_model.fit(X_8_train, y_train)
train_score_lr8 = lr8_model.score(X_8_train, y_train)
N = X_8_train.shape[0]
M = X_8_train.shape[1]
R2_korr = 1 - (1-train_score_lr8) * (N-1) / (N-M-1)
print("Lineare Regression 8. Ordnung - Training: R² = {:.2f}".format( R2_korr ))

##################################################################################
######Kreuzvalidierung zur Optimierung der Hyperparameter nicht vergessen!########
##################################################################################




alpha_candidates = np.logspace(-50, 50, 101) # üblicherweise verwendet man hier als Kandidaten eine logarithmische Spanne von Werten
ridge1_model = RidgeCV(alphas=alpha_candidates)
ridge1_model.fit(X_train, y_train)
alpha_opt = ridge1_model.alpha_
train_score_ridge1 = ridge1_model.score(X_train, y_train)
print("Ridge Regression 1. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge1))

ridge2_model = RidgeCV(alphas=alpha_candidates)
ridge2_model.fit(X_2_train, y_train)
alpha_opt = ridge2_model.alpha_
train_score_ridge2 = ridge2_model.score(X_2_train, y_train)
print("Ridge Regression 2. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge2))

ridge3_model = RidgeCV(alphas=alpha_candidates)
ridge3_model.fit(X_3_train, y_train)
alpha_opt = ridge3_model.alpha_
train_score_ridge3 = ridge3_model.score(X_3_train, y_train)
print("Ridge Regression 3. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge3))

ridge4_model = RidgeCV(alphas=alpha_candidates)
ridge4_model.fit(X_4_train, y_train)
alpha_opt = ridge4_model.alpha_
train_score_ridge4 = ridge4_model.score(X_4_train, y_train)
print("Ridge Regression 4. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge4))

ridge5_model = RidgeCV(alphas=alpha_candidates)
ridge5_model.fit(X_5_train, y_train)
alpha_opt = ridge5_model.alpha_
train_score_ridge5 = ridge5_model.score(X_5_train, y_train)
print("Ridge Regression 5. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge5))

ridge6_model = RidgeCV(alphas=alpha_candidates)
ridge6_model.fit(X_6_train, y_train)
alpha_opt = ridge6_model.alpha_
train_score_ridge6= ridge6_model.score(X_6_train, y_train)
print("Ridge Regression 6. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge6))

ridge7_model = RidgeCV(alphas=alpha_candidates)
ridge7_model.fit(X_7_train, y_train)
alpha_opt = ridge7_model.alpha_
train_score_ridge7= ridge7_model.score(X_7_train, y_train)
print("Ridge Regression 7. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge7))

ridge8_model = RidgeCV(alphas=alpha_candidates)
ridge8_model.fit(X_8_train, y_train)
alpha_opt = ridge8_model.alpha_
train_score_ridge8= ridge8_model.score(X_8_train, y_train)
print("Ridge Regression 8. Ordnung (alpha={:.2f}): Training R^2 = {:.2f}".format(alpha_opt, train_score_ridge8))




###################################################################################

# from matplotlib import pyplot as plt
# %matplotlib inline
# plt.hist(y_train)
# plt.xlabel("Value")
# plt.ylabel("Häufigkeit")


# labels, counts = np.unique(y_train, return_counts=True)
# # Finde die kleinste Häufigkeit
# min_counts = np.min(counts)
# # Finde das Label mit der kleinsten Häufigkeit
# rarest_label = labels[counts == min_counts]
# # Ausgabe des seltensten Labels und der entsprechenden Häufigkeit
# print("Das Label {} kommt nur {} Mal in den Trainingsdaten vor!".format(*rarest_label, min_counts))


##############################################################################################
# # Auswahl der zu testenden Hyperparameter (alle nicht spezifizierten bleiben auf dem Standardwert)
#################################################################################################
# param_grid = [
#     {'C': np.logspace(-4, 4, 9),        # Parameter C nach Vorlesung
#      'epsilon': np.logspace(-2, 2, 5),  # Parameter epsilon zur Steuerung 
#      'kernel': ['linear', 'rbf']},      # Verwendeter Kernel
# ]
# # Instanziieren des Objekts "GridSearchCV"
# svr_model = GridSearchCV(SVR(),            # Regressionsmodell, das benutzt werden soll
#                          param_grid,       # Auszuprobierenden Hyperparameterwerte
#                          scoring='r2',     # Zu verwendende Metrik
#                          cv=5,             # Anzahl der Folds für die Kreuvalidierung
#                          verbose=100,       # Menge des Output während der Suche (größere Zahl -> mehr Info)
#                          n_jobs=-1)        # Anzahl der parallel genutzen CPUs; -1: alle verfügbaren CPUs nutzen
# svr_model.fit(X_train, y_train)
# print(svr_model.best_params_)

#################################################
########## Trainieren Sie das Modell#############
#################################################

# Am Ende Modell mit optimalen Hyperparametern auf allen gelabelte Daten trainieren!
svr_model = SVR( C=100, epsilon=0.1, kernel='rbf')
svr_model.fit(X_train, y_train)
train_score_svr = svr_model.score(X_train, y_train)
print("Support Vector Regression: Training R^2 = {:.2f}".format(train_score_svr))

```
